```python
#one hot encoding using one hot encoder of scikit learn
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
#buidling a dummy employee dataset for example
data={'employee id':[1,2,3,4,5],'gender':['male','female','female','male','female'],'remarks':['good','bad','good','good','bad']}
df=pd.DataFrame(data)
print(f"employee data:\n{df}")
```

```python
#extract categorical columns from the datafarme
#here we extract the columns with object dataype as they are categorical columns
categorical_column=df.select_dtypes(include=['object']).columns.tolist()
```

```python
#initialize onehot encoder
encoder=OneHotEncoder(sparse_output=False)
#apply one hot encoding to the categorical columns
one_hot_encoded=encoder.fit_transform(df[categorical_column])
#create a dataframe with one hot encoded columns
one_hot_df=pd.DataFrame(one_hot_encoded,columns=encoder.get_feature_names_out(categorical_column))
#concatenate the one hot encoded data frame with original dataframe
df_encoded=pd.concat([df,one_hot_df],axis=1)
#drop the original categorical columns
df_encoded=df_encoded.drop(categorical_column,axis=1)
#print the encoded dataframe
print(f"encoded data:\n{df_encoded}")
```
