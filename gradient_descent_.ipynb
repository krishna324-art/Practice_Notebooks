```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#generate a simple dataset#y=2x+1
np.random.seed(42)
X=np.linspace(0,10,50)#50 evenly spaced points 0 and 10
true_m,true_c=2,1
y=true_m*X+true_c+np.random.normal(0,2,X.shape)#add somee noice
#y=true_m*X+true_c
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X,y,label="Data Points",color="blue")
plt.title("input dtaa")
plt.xlabel("X")
plt.ylabel("y")

plt.show()
```

```python
#initialize the parameter(slope and intercept)
m,c=0.0,0.0#starting with random guesses
learning_rate=0.01#learning rate
iterations=1000#number of iterations
```

```python
#store loss for plotting
loss_history=[]
```

```python
for i in range(iterations):
  #predicted y values
  y_pred=m*X+c

  #compute gradients
  m_grad=(-2/len(X))*np.sum(X*(y-y_pred))
  c_grad=(-2/len(X))*np.sum(y-y_pred)

  #update parameters
  m=m-learning_rate*m_grad
  c=c-learning_rate*c_grad

  #compute loss(MSE)
  loss=np.mean((y-y_pred)**2)
  loss_history.append(loss)

  #print updates every 100 iterations
  if i%10==0:
    print(f"Iteration {i}: Loss={loss:.4f}, m={m:.4f}, c={c:.4f}")
```

```python
#print final parameters
print(f"Final parameters:m={m:4f},c={c:.4f}")
#plot the dataset and fitted line
plt.figure(figsize=(10,6))
plt.scatter(X,y,label="Data Points",color="blue")
plt.plot(X,m*X+c,label=f"Fitted Line:y={m:.2f}X+{c:.2f}",color="red")
plt.title("Gradient Descent:line fitting")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
